# **PELABELAN DATA DENGAN METODE BERT **

```python
import pandas as pd
from transformers import pipeline

# Membaca file Excel
file_path = '/content/sample_data/Data_Twitter_Combined 100824.xlsx'
df = pd.read_excel(file_path)

# Memuat pipeline BERT untuk analisis sentimen
classifier = pipeline('sentiment-analysis')

# Mendefinisikan daftar kata kunci untuk setiap kategori
positif_keywords = ['dukung', 'protes', 'menentang', 'menolak', 'semangat', 'mantap', 'hebat', 'oke', 'bagus', 'senang', 'puas', 'favorit']
negatif_keywords = ['benci', 'jelek', 'buruk', 'marah', 'kesal', 'rugi', 'nyesel', 'tidak suka', 'kecewa', 'sampah']
netral_keywords = ['saya', 'ini', 'adalah', 'mengenai', 'tentang', 'dari', 'ke', 'dengan', 'untuk']

def custom_classification(text):
    text_lower = text.lower()

    # Cek apakah teks mengandung kata kunci dari setiap kategori
    if any(keyword in text_lower for keyword in positif_keywords):
        return 'Positif'
    elif any(keyword in text_lower for keyword in negatif_keywords):
        return 'Negatif'
    elif any(keyword in text_lower for keyword in netral_keywords):
        return 'Netral'
    else:
        # Gunakan model BERT sebagai fallback
        result = classifier(text)[0]['label']
        # Periksa label dari BERT
        if result == 'LABEL_0':  # Misalkan LABEL_0 adalah 'NEGATIVE'
            return 'Negatif'
        elif result == 'LABEL_1':  # Misalkan LABEL_1 adalah 'POSITIVE'
            return 'Positif'
        else:
            return 'Netral'

# Melakukan klasifikasi pada kolom full_text
df['sentiment'] = df['full_text'].apply(custom_classification)

# Menampilkan 5 baris pertama dari hasil sentimen
print(df[['full_text', 'sentiment']].head())

# Menyimpan hasil ke file Excel baru
output_file_path = '/content/sample_data/Data_Twitter_Combined_Analyzed_BERT.xlsx'
df.to_excel(output_file_path, index=False)
```

```python
import pandas as pd
import re
from nltk.tokenize import word_tokenize
import nltk

# Mengunduh resource untuk tokenisasi dari NLTK
nltk.download('punkt')

# Membaca file Excel
file_path = '/content/sample_data/Data_Twitter_Combined 100824.xlsx'
df = pd.read_excel(file_path)

# Melakukan case folding pada kolom full_text
df['full_text'] = df['full_text'].str.lower()

# Fungsi untuk membersihkan teks
def clean_text(text):
    # Menghapus URL
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    # Menghapus karakter khusus, angka, dan ekstra spasi
    text = re.sub(r'[^a-z\s]', '', text)
    # Menghapus spasi ekstra
    text = ' '.join(text.split())
    return text

# Membuat kolom baru untuk teks yang telah dibersihkan
df['cleaned_text'] = df['full_text'].apply(clean_text)

# Kamus Normalisasi
normalization_dict = {
    'gw': 'gue',
    'loe': 'lo',
    'yg': 'yang',
    'gk': 'gak',
    'bgt': 'banget',
    'klo': 'kalau',
    'tdk': 'tidak',
    'aj': 'aja',
    'k': 'ke',
    'd': 'de'
}

# Fungsi untuk normalisasi teks
def normalize_text(text):
    tokens = text.split()
    normalized_tokens = [normalization_dict.get(token, token) for token in tokens]
    return ' '.join(normalized_tokens)

# Normalisasi kolom cleaned_text
df['normalized_text'] = df['cleaned_text'].apply(normalize_text)

# Fungsi untuk tokenisasi
def tokenize_text(text):
    return word_tokenize(text)

# Melakukan tokenisasi pada kolom normalized_text
df['tokenized_text'] = df['normalized_text'].apply(tokenize_text)

# Menampilkan 5 baris pertama dengan kolom normalized_text dan tokenized_text
print(df[['normalized_text', 'tokenized_text']].head())

# Menyimpan hasil ke file Excel baru
output_file_path = '/content/sample_data/Data_Twitter_Combined_Normalized_Tokenized.xlsx'
df.to_excel(output_file_path, index=False)

# Menampilkan lokasi file output
print(f"File telah disimpan di: {output_file_path}")
```

## **Preprocessing STOPWORD Removal**

```python
import pandas as pd
import re
from nltk.tokenize import word_tokenize
import nltk
import os

# Mengunduh resource untuk tokenisasi dari NLTK
nltk.download('punkt')

# Membaca file Excel
file_path = '/content/sample_data/Data_Twitter_Combined 100824.xlsx'
df = pd.read_excel(file_path)

# Melakukan case folding pada kolom full_text
df['full_text'] = df['full_text'].str.lower()

# Fungsi untuk membersihkan teks
def clean_text(text):
    # Menghapus URL
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    # Menghapus karakter khusus, angka, dan ekstra spasi
    text = re.sub(r'[^a-z\s]', '', text)
    # Menghapus spasi ekstra
    text = ' '.join(text.split())
    return text

# Membuat kolom baru untuk teks yang telah dibersihkan
df['cleaned_text'] = df['full_text'].apply(clean_text)

# Kamus Normalisasi
normalization_dict = {
    'gw': 'gue',
    'loe': 'lo',
    'yg': 'yang',
    'gk': 'gak',
    'bgt': 'banget',
    'klo': 'kalau',
    'tdk': 'tidak',
    'aj': 'aja',
    'k': 'ke',
    'd': 'de'
}

# Fungsi untuk normalisasi teks
def normalize_text(text):
    tokens = text.split()
    normalized_tokens = [normalization_dict.get(token, token) for token in tokens]
    return ' '.join(normalized_tokens)

# Normalisasi kolom cleaned_text
df['normalized_text'] = df['cleaned_text'].apply(normalize_text)

# Tokenisasi
def tokenize_text(text):
    return word_tokenize(text)

df['tokenized_text'] = df['normalized_text'].apply(tokenize_text)

# Mengunduh atau membuat kamus stopword
stopword_file_path = '/content/sample_data/stopwords.txt'

# Jika file stopwords tidak ada, buat file contoh
if not os.path.isfile(stopword_file_path):
    stopwords = [
        'ada', 'adapun', 'agak', 'agaknya', 'begini', 'beginian', 'cara',
        'caranya', 'cukup', 'cukupkah', 'dahulu', 'dalam', 'dan', 'dapat',
        'dari', 'daripada', 'datang', 'dekat', 'entah', 'enggak', 'guna',
        'gunakan', 'hal', 'hanya', 'kok'
    ]
    with open(stopword_file_path, 'w') as file:
        file.write('\n'.join(stopwords))

# Membaca kamus stopword
with open(stopword_file_path, 'r') as file:
    stopwords = set(file.read().split())

# Fungsi untuk menghapus stopwords
def remove_stopwords(tokens):
    return [token for token in tokens if token not in stopwords]

# Menghapus stopwords dari kolom tokenized_text
df['filtered_text'] = df['tokenized_text'].apply(remove_stopwords)

# Menggabungkan token menjadi teks yang sudah difilter
df['filtered_text'] = df['filtered_text'].apply(lambda tokens: ' '.join(tokens))

# Menampilkan 5 baris pertama dengan kolom filtered_text
print(df[['filtered_text']].head())

# Menyimpan hasil ke file Excel baru
output_file_path = '/content/sample_data/Data_Twitter_Combined_Stopwords_Removed.xlsx'
df.to_excel(output_file_path, index=False)

# Menampilkan lokasi file output
print(f"File telah disimpan di: {output_file_path}")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
from imblearn.over_sampling import SMOTE

# Membaca file Excel
file_path = '/content/sample_data/Data_Twitter_Combined_Stemmed.xlsx'
df = pd.read_excel(file_path)

# Tambahkan kolom 'label' berdasarkan aturan klasifikasi
def assign_label(text):
    if 'boikot' in text or 'protes' in text or 'menentang' in text or 'menolak' in text or 'dukung Palestina' in text:
        return 1  # Positif
    elif 'senang' in text or 'bagus' in text or 'favorit' in text or 'setuju' in text:
        return 0  # Negatif
    else:
        return 2  # Netral

df['label'] = df['stemmed_text'].apply(assign_label)

# Fitur dan target
X = df['stemmed_text']
y = df['label']

# TF-IDF Vectorization
tfidf_vectorizer = TfidfVectorizer()
X_tfidf = tfidf_vectorizer.fit_transform(X)

# Split data
X_train_tfidf, X_test_tfidf, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.3, random_state=42)

# Apply SMOTE
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_tfidf, y_train)

# Train model
nb_model = MultinomialNB()
nb_model.fit(X_train_resampled, y_train_resampled)

# Predictions
y_pred = nb_model.predict(X_test_tfidf)

# Evaluation
print(classification_report(y_test, y_pred, target_names=['Negatif', 'Positif', 'Netral']))
```
